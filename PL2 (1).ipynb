```python
import yfinance as yf
import pandas as pd
```

```python
# list stocks to use - replace this with a list of all stocks eventually
input = ['AAPL', 'COST', 'TSLA', 'SAVE']
```

```python
# create an empty dataframe to add stock info to
columns = ['Ticker', 'Name', 'Industry', 'PEG', 'Price']
df = pd.DataFrame(columns=columns)
```

```python
# create a function to get info from every stock
def getInfo(stock_list):
    for stock in stock_list: 
        ticker = yf.Ticker(stock)
        info = ticker.info
        industry = info['industry']
        sector = info['sector']
        price = info['currentPrice']
        peg = info['pegRatio']
        name = info['shortName']

        # add stock to the dataframe
        df.loc[len(df.index)] = [stock, name, industry, peg, price]

getInfo(input)
df
```

```python
# write a function to evaluate if the peg ratio is good, based on industry

def evaluatePEG(stock_list):
    # define an empty dictionary to add peg evals to
    evals = {}

    # evaluate if the peg is good, based on the industry
    for stock in stock_list:

        # get the relevant info
        ticker = yf.Ticker(stock)
        info = ticker.info
        industry = info['industry']
        peg = info['pegRatio']

        # set peg eval to good by default
        peg_eval = 'Good'

        # see if the peg eval should be changed to bad based on industry
        if industry == 'Consumer Electronics' and peg > 1:
            peg_eval =' Bad'
        if industry == 'Discount Stores' and peg > 6: 
            peg_eval = 'Bad'
        if industry == 'Auto Manufacturers' and peg > 30:
            peg_eval = 'Bad'
        if industry == 'Airlines' and peg > 1: 
            peg_eval = 'Bad'

        # add the stock (key) and the peg eval (value) to the evals dictionary created above
        evals[stock] = peg_eval

    # add the peg eval column to the df by mapping it to the stock name
    df['PEG Eval'] = df['Ticker'].map(evals)
        
evaluatePEG(input)
df
```
